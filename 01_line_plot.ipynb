## Adding limits

```python
import seaborn as sns
import matplotlib.pyplot as plt

# load data set
flower=sns.load_dataset("iris")
flower

# draw a line plot 
sns.lineplot(x="sepal_length",y="sepal_width",data=flower)
plt.title("Flowers Plot")
plt.xlim(2)  # for x axis
plt.ylim(1)  # for y axis
```

# Size of Figure

```python
import seaborn as sns
# we need a libraries to import for estimator 
from numpy import mean
import matplotlib.pyplot as plt
sns.set_style(style=None, rc=None)
# load data set
flower=sns.load_dataset("iris")
#change figure
plt.figure(figsize=(7,6))

# draw a line plot 
sns.barplot(x="sepal_length",y="sepal_width",data=flower,estimator=mean)
plt.title("Flowers Plot")

# Style adding 
sns.set_style("dark")


```
