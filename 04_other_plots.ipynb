```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

dots= sns.load_dataset("dots")

# defining a color pelette
pelette = sns.color_palette('rocket_r') 
# now a line plot but with a new name i.e relplot
sns.relplot(data=dots,x="time",y="firing_rate",hue="coherence", col="align",
            size="choice",kind="line",size_order=["T1","T2"],
            palette=pelette,aspect=.75,
            facet_kws=dict(sharex=False),
           )
```

```python
import seaborn as sns
sns.set_theme(style="ticks",palette="pastel")
tips=sns.load_dataset("tips")
sns.boxplot(data=tips,x="day",y="total_bill",hue="smoker",palette=["m","g"])
sns.despine(offset=10,trim=True)
```

# violin plot

```python
import seaborn as sns
sns.set_theme(style="whitegrid")
tips=sns.load_dataset("tips")

sns.violinplot(data=tips,x="day",y="total_bill",hue="smoker",split=True,
               inner="quart",linewidth=1,palette={"Yes":"b","No":".85"})
sns.despine(left=True)
```

# Scatter  plot

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="whitegrid")

diamonds= sns.load_dataset("diamonds")

f , ax=plt.subplots(figsize=(6.5,6.5))
sns.despine(f,left=True,bottom=True)
clarity_ranking=["I1","SI2","SI1","VS2","VS1","VVS2","VVS1","IF"]
sns.scatterplot(data=diamonds,x="carat",y="price",
               palette="ch:r=-.2,d=.3_r",hue="clarity",
               size="depth",hue_order=clarity_ranking,
               sizes=(1,8),linewidth=0,ax=ax)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="ticks")
# initiative the figure with a logarithmic x axis
f,ax=plt.subplots(figsize=(7,6))
ax.set_xscale("log")

planets= sns.load_dataset("planets")

sns.boxplot(data=planets, x="distance",y="method",
           whis=[0,100],width=.6,palette="vlag")

# add in points to show each observation
# stripplot dont do anything but mark something on the given plots Below graph has black kinda strip plot
sns.stripplot(data=planets,x="distance",y="method",
             size=4,color=".3",linewidth=0)

# now we need to tweak the viusual presentation
ax.xaxis.grid(True)
ax.set(ylabel="")
sns.despine(trim=True,left=True)
```

```python
import seaborn as sns
sns.set_theme(style="ticks")

planets=sns.load_dataset("planets")
g=sns.JointGrid(data=planets,x="year",y="distance",marginal_ticks=True)

# setting a log on Y axis
g.ax_joint.set(yscale="log")

cax=g.figure.add_axes([.15,.55,.02,.2])

g.plot_joint(
    sns.histplot,discrete=(True,False),
    cmap="light:#03012d",pmax=.8,cbar=True,cbar_ax=cax
)
g.plot_marginals(sns.histplot,element="step",color="#03012d")
```

```python
# Boxen plot

import seaborn as sns
sns.set_theme(style="whitegrid")

diamonds=sns.load_dataset("diamonds")

clarity_ranking=["I1","SI2","SI1","VS2","VS1","VVS2","VVS1","IF"]

sns.boxenplot(data=diamonds,x="clarity",y="carat",
             color="b",order=clarity_ranking,scale="linear")
```

```python
import seaborn as sns
sns.set_theme(style="ticks")

penguins=sns.load_dataset("penguins")
sns.pairplot(penguins,hue="species")
```

```python

import seaborn as sns
sns.set_theme(style="whitegrid")

planets=sns.load_dataset("planets")

cmap=sns.cubehelix_palette(rot=.2,as_cmap=True)
g= sns.relplot(data=planets,x="distance",y="orbital_period",hue="year",size="mass",palette=cmap,sizes=(10,200))

g.set(xscale="log",yscale="log")
g.ax.xaxis.grid(True,"minor",linewidth=.25)
g.ax.yaxis.grid(True,"minor",linewidth=.25)
g.despine(left=True,bottom=True)
```
